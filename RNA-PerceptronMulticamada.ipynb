 MLP - Perceptron Multicamada

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import warnings
import sklearn.exceptions
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv('winequality-red.csv',sep=';')
df.head()
```

```python
df.describe().transpose() # mostra a estrutura estatística dos dados
```

```python
X = df.drop('quality',axis=1) # esse comando retira a coluna quality
X
```

```python
y = df['quality']
y
```

```python
#separando em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.333)
X_train, X_test
```

```python
#escalonandor
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
#escanolando
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
X_train, X_test
```

```python
# 2 camadas ocultas; 11 neurônios cada; função de ativação logistica
#número máximo de interações 15000; taxa de aprendizado adaptativa; inicial 0.5
#taxa de momento: 0.1
# algoritmo de treinamento: descida de gradiente estocástico (sgd)
mlp = MLPClassifier(hidden_layer_sizes=(11,11), activation='logistic', max_iter = 15000,
                    alpha=1e-6, solver= 'sgd', verbose=10, tol=1e-9, random_state=1, 
                    learning_rate='adaptive', momentum=0.1, learning_rate_init=0.5)
```

```python
mlp.fit(X_train, y_train)
```

```python
predictions = mlp.predict(X_test)
predictions
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
# acurácia 60%
# média ponderada da precisão 57%
# recall 60%; f1-score 58%
warnings.filterwarnings("ignore", category=sklearn.exceptions.UndefinedMetricWarning)
print(classification_report(y_test, predictions))
```

```python
#  diferença entre o conjunto predito e as classes que constavam previamente no dataset
e = y_test - predictions
for item in e.iteritems():
    if(item[1] != 0):
        print(item)
```
